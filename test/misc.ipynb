```python
from uilc.utils.misc import extend_signal, half_ceil
```

```python
time = np.linspace(0, 2*np.pi, 300, endpoint=True)
sig = np.sin(time)
```

```python
plt.plot(time, sig)
```

```python
period = 1.5*np.pi
```

```python
sig_ext, time_ext = extend_signal(
    sig, time, 5, 
    central=False, period= period,
    edge_matching=True)
```

```python
fig = plt.figure()
ax = fig.add_subplot(2, 1, 1)
ax2 = fig.add_subplot(2,1, 2)

ax.plot(time, sig)
ax.axvline(x=period)
ax2.plot(time_ext, sig_ext)
plt.show()
```
